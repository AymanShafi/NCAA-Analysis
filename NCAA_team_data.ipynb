```python
from selenium import webdriver
import pandas as pd
driver = webdriver.Chrome(executable_path='C:\\Users\\GV62\\Downloads\\chromedriver_win32_new\\chromedriver.exe')
final_df = pd.DataFrame()
for i in range(2011,2020):
    driver.get("https://www.sports-reference.com/cbb/seasons/"+str(i)+"-school-stats.html")
    t1 = driver.find_element_by_id("basic_school_stats").get_attribute("outerHTML")
    df1 = pd.DataFrame()
    df1 = pd.read_html(t1)[0]
    driver.get("https://www.sports-reference.com/cbb/seasons/"+str(i)+"-advanced-school-stats.html")
    t2 = driver.find_element_by_id("adv_school_stats").get_attribute("outerHTML")
    df2 = pd.DataFrame()
    df2 = pd.read_html(t2)[0] 
    df1.columns = df1.columns.droplevel(0)
    df2.columns = df2.columns.droplevel(0)
    df1 = df1.drop(['Unnamed: 16_level_1'],axis=1)
    df2 = df2.drop(['Unnamed: 16_level_1'],axis=1)
    df1 = df1.dropna()
    df2 = df2.dropna()
    cols = [i for i in range(0,16)]
    df2.drop(df2.columns[cols],axis=1,inplace=True)
    result = pd.concat([df1, df2], axis=1, sort=False)
    lst = ['NCAA']
    result = result[result['School'].str.contains("NCAA", case = False)]
    result['Year'] = i
    final_df = pd.concat([final_df,result])
driver.close()
final_df
```

```python
final_df.to_csv('NCAA_Team_Data.csv')
```

```python
final_df.columns = ['Rk', 'School', 'G', 'W', 'L', 'W-L%', 'SRS', 'SOS', 'W_conf', 'L_conf', 'W_home', 'L_home',
       'W_away', 'L_away', 'Tm.', 'Opp.', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%',
       'FT', 'FTA', 'FT%', 'ORB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'Pace', 'ORtg', 'FTr', '3PAr', 'TS%', 'TRB%', 'AST%', 'STL%', 'BLK%',
       'eFG%', 'TOV%', 'ORB%', 'FT/FGA', 'Year']
```
